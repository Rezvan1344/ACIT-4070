```python
import numpy as np
from matplotlib import pyplot as plt
```

```python

if __name__ == '__main__':
    # to make randoms same each time
    # np.random.seed(1)
    N = 100

    eps_max = 0.2
    eps_min = 0
    eps_step = 0.005

    EPSILONS = np.arange(eps_min, eps_max, eps_step)
    print(f"{EPSILONS}\n")
    fig, ax = plt.subplots()

    for n in [10, 20, 50, 100, 1000]:
        X = [np.random.uniform(1.50, 2.00) for _ in range(n)]
        # X = [round(np.random.uniform(1.50, 2.00), 2) for _ in range(n)]
        avg_X = np.average(X)

        print(f"{X}\n")
        print(f"{avg_X}\n")

        ERRORS = []
        avg_100Xis = []
        print("----- Enter Epsilon loop -----")
        for eps in EPSILONS:
            print(f"{eps}")
            error_N = []
            for i in range(N):
                print(f"----- N={i} -----")

                # Add (Epsilon * r) to the X
                Xi = [x + np.random.uniform(-eps, eps) for x in X]
                print(f"Xi = X + (eps*r) =\n{Xi}")

                # Average of Xi + (eps * r )
                avg_Xi = np.average(Xi)
                print(f"{avg_X}")
                print(f"{avg_Xi}")

                error = abs(avg_X - avg_Xi) * 100
                print(f"{error}\n")
                error_N.append(error)
                # avg_100Xis.append(avg_Xi)
            error_avg = np.average(error_N)
            print(f"{error_avg}")

            # error_Xi = abs(avg_X - total_avg_Xi)
            ERRORS.append(error_avg)
            print(f"{ERRORS}")
            print()

        ax.plot(EPSILONS, ERRORS, label=n)

    plt.xlabel('Epsilons')
    plt.ylabel('Errors in %')
    plt.legend()
    plt.savefig(F'fig_with_loop_{N}-esp{eps_min}-{eps_max}.png')
    plt.show()
```
